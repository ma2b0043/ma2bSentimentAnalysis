```python
import graphlab
```

# Reading some product review data

```python
products = graphlab.SFrame('amazon_baby.csv')
```

# Exploring the dataset

```python
products.head()
```

# Build word count vector for each review

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
products['name'].show()
```

# Exploring Vuli Sophie

```python
giraffe_reviews = products[products['name'] == 'Vulli Sophie the Giraffe Teether']
```

```python
giraffe_reviews['rating'].show(view='Categorical')
```

# Build a Sentiment Classifer

```python
#Define a Posituve and Negative Review
```

```python
#igonre 3* star review
giraffe_reviews = products[products['rating'] != '3']
```

```python
#positive reviews are 4* and 5*
products['sentiment'] = products['rating'] >= 4
```

```python
products.head()
```

# Training the sentiment analysis

```python
#Spliting the data into train data and test data
train_data,test_data = products.random_split(0.8,seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,target='sentiment',features=['word_count'],validation_set=test_data)
```

# Evaluate the sentiment model

```python
sentiment_model.evaluate(test_data, metric='roc_curve')#roc curver is used to explore false positive and false neagtives
```

```python
sentiment_model.show(view='Evaluation')
```

# Apply learned model to apply sentiment for giraffe

```python
giraffe_reviews['predicted_sentiment'] = sentiment_model.predict(giraffe_reviews, output_type='probability')
```

```python
giraffe_reviews.head()
```

# Sorting the reviews based on predicted_sentiments

```python
giraffe_reviews = giraffe_reviews.sort('predicted_sentiment', ascending=False)#sort the reviews in descending on basis of predicted sentiment
```
